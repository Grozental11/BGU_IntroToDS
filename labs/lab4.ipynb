```python
data = pd.read_csv('https://raw.githubusercontent.com/DanielBa125/DsAssignment5/main/hero_network.csv?token=GHSAT0AAAAAACNZ3XR5QZWJHSYWBXOCC4QSZOA2CSQ')
```

```python
#Q1
all_heroes = pd.concat([data['hero1'], data['hero2']])
hero_counts = all_heroes.value_counts()

single_friends_count = (hero_counts == 1).sum()
print(f"{single_friends_count} characters have exactly one friend")
```

```python
#Q2
captain_america_friends = set(all_heroes[(data['hero1'] == 'CAPTAIN AMERICA') | (data['hero2'] == 'CAPTAIN AMERICA')])
spiderman_friends = set(all_heroes[(data['hero1'] == 'SPIDER-MAN/PETER PAR') | (data['hero2'] == 'SPIDER-MAN/PETER PAR')])
ironman_friends = set(all_heroes[(data['hero1'] == 'IRON MAN/TONY STARK ') | (data['hero2'] == 'IRON MAN/TONY STARK ')])

captain_america_friends.discard('CAPTAIN AMERICA')
spiderman_friends.discard('SPIDER-MAN/PETER PAR')
ironman_friends.discard('IRON MAN/TONY STARK ')

mutual_friends = captain_america_friends & spiderman_friends & ironman_friends
mutual_friends_count = len(mutual_friends)
print(f"The number of mutual friends of Captain America, Spider-Man and Iron-Man is {mutual_friends_count}")
```

```python
#Q3
unique_heroes = pd.unique(all_heroes)
longest_hero_name = max(unique_heroes, key=len)
shortest_hero_name = min(unique_heroes, key=len)

(longest_hero_name, shortest_hero_name)
```

```python
#Q4
def not_cap_friends(hero_name):
    hero_friends = set(all_heroes[(data['hero1'] == hero_name) | (data['hero2'] == hero_name)])
    captain_america_friends = set(all_heroes[(data['hero1'] == 'CAPTAIN AMERICA') | (data['hero2'] == 'CAPTAIN AMERICA')])
    not_captain_america_friends = hero_friends.difference(captain_america_friends)
    return not_captain_america_friends

not_cap_friends("SPIDER-MAN/PETER PAR")
```

```python
#Q5
spiderman_rows = data[(data['hero1'] == 'SPIDER-MAN/PETER PAR') | (data['hero2'] == 'SPIDER-MAN/PETER PAR')]
spiderman_friends = set(spiderman_rows['hero1']).union(set(spiderman_rows['hero2']))
spiderman_friends.discard('SPIDER-MAN/PETER PAR')
friends_data = data[data['hero1'].isin(spiderman_friends) | data['hero2'].isin(spiderman_friends)]
friends_of_friends = set(friends_data['hero1']).union(set(friends_data['hero2']))
friends_of_friends.discard('SPIDER-MAN/PETER PAR')
friends_of_friends = friends_of_friends - spiderman_friends
number_of_unique_friends = len(friends_of_friends)

number_of_unique_friends
```

```python
url = 'https://raw.githubusercontent.com/DanielBa125/DsAssignment5/main/football_results.csv'
data = pd.read_csv(url)
```

```python
#Q1
def max_wins_away(team_name):
    home_games = data[data['home_team'] == team_name]
    away_games_against_team = home_games[['away_team', 'away_score']]
    scores_by_team = away_games_against_team.groupby('away_team').sum()
    top_opponent = scores_by_team['away_score'].idxmax()

    return top_opponent

print(f"Israel -> '{max_wins_away('Israel')}'\n Italy -> '{max_wins_away('Italy')}'\n Argentina -> '{max_wins_away('Argentina')}'\n")
```

```python
#Q2
def teams_history(team1, team2):
    data = pd.read_csv(url)
    filtered_games = data[((data['home_team'] == team1) & (data['away_team'] == team2)) |
                          ((data['home_team'] == team2) & (data['away_team'] == team1))].copy()
    filtered_games['total_goals'] = filtered_games['home_score'] + filtered_games['away_score']

    result_string = ""

    for home_team in [team1, team2]:
        home_games = filtered_games[filtered_games['home_team'] == home_team]
        if home_games.empty:
            result_part = f'{home_team} at home: No games found. '
        else:
            max_goals_game = home_games.loc[home_games['total_goals'].idxmax()]
            min_goals_game = home_games.loc[home_games['total_goals'].idxmin()]
            result_part = (
                f'{home_team} at home: '
                f"Most Goals - Date: {max_goals_game['date']}, Result: {max_goals_game['home_team']} {max_goals_game['home_score']} - {max_goals_game['away_team']} {max_goals_game['away_score']}"
                f"Fewest Goals - Date: {min_goals_game['date']}, Result: {min_goals_game['home_team']} {min_goals_game['home_score']} - {min_goals_game['away_team']} {min_goals_game['away_score']}; "
            )

        result_string += result_part

    return result_string
teams_history = teams_history('Germany', 'Spain')
teams_history
```
